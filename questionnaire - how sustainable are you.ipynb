```python
#titles
text = "How sustainable are you?"
title_text = text.title()
print(title_text)
text = 'A questionnaire on how sustainable you are by determining your sustainability score!'
print(text)

print("\n")


total_emissions = 0

average_emissions = 7250000

def calculate_sustainability_score(total):
    score = max(1, min(10, 10 - (total / (average_emissions / 10))))
    return score

# Question 1
def open_question1():
    global total_emissions
    answer = input("1. How many hours a year do you fly in a plane? (Enter the amount of hours) ")
    try:
        numeric_answer = float(answer)
        points1 = numeric_answer * 90000
        total_emissions += points1 
        print(f"You flew for {numeric_answer} hours and this emits on average {points1} grams of CO2.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

open_question1()
print("\n")

# Question 2
question = "2. What is your diet? (obviously loads of specific diets are not included, please choose the diet closest to yours)"
options = {
    '1': ("Vegetarian", 1700000),
    '2': ("Vegan", 1500000),
    '3': ("Average", 2500000),
    '4': ("Meat lover", 3300000)
}

def ask_question():
    print(question)
    for key, value in options.items():
        print(f"{key}. {value[0]}")

def get_points():
    global total_emissions
    ask_question()
    user_choice = input("Please select an option (1-4): ")

   
    if user_choice in options:
        selected_option = options[user_choice]
        total_emissions += selected_option[1] 
        print(f"Your diet ({selected_option[0]}) emits on average {selected_option[1]} grams of CO2 annually.")
    else:
        print("Invalid choice. Please choose a valid option.")

get_points()
print("\n") 

# Question 3
def open_question3():
    global total_emissions
    answer = input("4.	How much gas do you use a year? (in kwh)")
    try:
        numeric_answer = float(answer)
        points3 = numeric_answer * 185
        total_emissions += points3 
        print(f"You have used {numeric_answer} kWh of gas this year and this emits on average {points3} grams of CO2.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

open_question3()
print("\n")

# Question 4
def open_question4():
    global total_emissions  
    answer = input("4. How many kilometers do you travel per car in a year? (fill in number in km – zero if you travel with an electric car) ")
    try:
        numeric_answer = float(answer)
        points4 = numeric_answer * 95
        total_emissions += points4 
        print(f"You drove {numeric_answer} kilometers this year and this emitted on average {points4} grams of CO2.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

open_question4()
print("\n")  

sustainability_score = calculate_sustainability_score(total_emissions)
print(f"Your total sustainability score (CO2 emissions) is: {total_emissions} grams of CO2.")
print(f"Your sustainability score on a scale of 1 to 10 is: {sustainability_score:.2f}")

print("\n")  

print("\nThis is the end of this questionnaire, thank you for filling it in. Hopefully this questionnaire has given you some insight into you anual CO2 emissions and areas where you can be more sustainable. Let's all do something for a greener and more sustainable future! ")

```
